```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Calculate average error of prediction

dataset = np.loadtxt(open("dataPredictions.csv"), delimiter=',')

accSum = 0
for i in range(int(len(dataset) / 2)):
    accSum += sum(100.0 * abs(dataset[2 * i] - dataset[(2 * i) + 1]) / dataset[(2 * i) + 1])
print ("Average Error: %.2f" % (accSum / (2.0 * len(dataset))))
```

```python
# Plot predicted vs. real values of confirmed cases for each country

realData = dict()
predData = dict()

for country in ["Italy", "Spain", "USA", "Brazil"]:
    realData[country] = []
    predData[country] = []
    
for i in range(int(len(dataset) / 2)):
    predData["Italy"].append(dataset[2 * i][0])
    realData["Italy"].append(dataset[(2 * i) + 1][0])
    
    predData["Spain"].append(dataset[2 * i][1])
    realData["Spain"].append(dataset[(2 * i) + 1][1])
    
    predData["USA"].append(dataset[2 * i][2])
    realData["USA"].append(dataset[(2 * i) + 1][2])
    
    predData["Brazil"].append(dataset[2 * i][3])
    realData["Brazil"].append(dataset[(2 * i) + 1][3])
    
xlabel = ["Mar 22", "Mar 23", "Mar 24", "Mar 25", "Mar 26", "Mar 27", "Mar 28", "Mar 29", "Mar 30", "Mar 31", "Apr 1"]

for country in ["Italy", "Spain", "USA", "Brazil"]:
    plt.plot(realData[country], label=country + "_real")
    plt.plot(predData[country], label=country+"_predicted")
    plt.legend()
    plt.xticks(range(len(xlabel)), xlabel, rotation=70)
    plt.show()
```
